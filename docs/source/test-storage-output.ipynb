# Reading General Datasets from S3 Buckets

## Reading MERRA NetCDF directly

```python
import xarray as xr
merra = xr.open_dataset("./my-private-bucket/MERRA2_401.tavg1_2d_flx_Nx.20200916.nc4")
conus = merra.sel(lat=slice(31, 46), lon=slice(-116, -69))
conus.EVAP.mean(["lat", "lon"]).plot()
```
